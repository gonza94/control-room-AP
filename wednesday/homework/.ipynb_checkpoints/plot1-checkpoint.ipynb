```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'
from scipy.optimize import curve_fit
```

```python
data = np.loadtxt('data2.csv',delimiter=',')
```

```python
def cosinefunc(x,phase_offset,amp,avg_val):
    phase = (np.pi/180.)*x
    phase_offset = phase_offset*(np.pi/180.)
    y = amp*np.cos(phase + phase_offset) + avg_val
    return y
```

```python
popt, pcov = curve_fit(cosinefunc, data[:,0], data[:,1], p0=[0.0,np.max(data[:,1])/2,np.mean(data[:,1])])
print ("(phase_offset,amp,avg_val) = ",(popt))
```

```python
fig,ax = plt.subplots(1,1,figsize = (8,6))

ax.scatter(data[:,0],data[:,1],label = 'pyOrbit')

xfit = np.linspace(0,360,100)
yfit = cosinefunc(xfit,*popt)

ax.plot(xfit,yfit, label = 'Cosine Fit', color = 'cyan')

ax.set_xlabel('Cavity Phase [deg]',fontsize = 30)
ax.set_ylabel('BPM Phase [deg]',fontsize = 30)

ax.tick_params(axis='both',labelsize=24)

ax.legend(fontsize = 20)

plt.show()
plt.close()
```

```python
e_kin_ini = 0.0025 # in [GeV]
mass = 0.939294    # in [GeV]
gamma = (mass + e_kin_ini)/mass
beta = np.sqrt(gamma*gamma - 1.0)/gamma
print ("relat. gamma=",gamma)
print ("relat.  beta=",beta)
bpm_frequency = 805.0e+6 # MHz
v_light = 2.99792458e+8  # in [m/sec]
s = 0.25 #m
L_dist =0.25
```

```python
data_rad = data*np.pi/180
```

```python
dw = mass*beta*((L_dist*2*np.pi*bpm_frequency/v_light)*(np.divide(1,data_rad[:,1]+popt[0]*np.pi/180))-beta)
```

```python
qV = (popt[1]*np.pi/180)*mass*beta**3*v_light/(2*np.pi*bpm_frequency*L_dist)
```

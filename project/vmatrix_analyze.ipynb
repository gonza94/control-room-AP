```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'

from matplotlib import cbook, cm
from matplotlib.colors import LightSource
```

```python
DCV_VA=['DTL_Mag:PS_DCV621',
'CCL_Mag:PS_DCV103',
'CCL_Mag:PS_DCV105',
'CCL_Mag:PS_DCV109',
'CCL_Mag:PS_DCV111',
'CCL_Mag:PS_DCV203',
'CCL_Mag:PS_DCV205',
'CCL_Mag:PS_DCV209',
'CCL_Mag:PS_DCV211',
'CCL_Mag:PS_DCV303',
'CCL_Mag:PS_DCV305',
'CCL_Mag:PS_DCV309',
'CCL_Mag:PS_DCV311',
'CCL_Mag:PS_DCV401',
'CCL_Mag:PS_DCV403',
'CCL_Mag:PS_DCV405',
'CCL_Mag:PS_DCV407',
'CCL_Mag:PS_DCV409',
'CCL_Mag:PS_DCV411']

DCV_VA = [s[-6:] for s in DCV_VA]
```

```python
vmat = np.loadtxt('orm_y.dat')

x = np.arange(0, vmat.shape[1])+1
y = np.arange(0, vmat.shape[0])+1
x, y = np.meshgrid(x, y)

bpmpositions = np.loadtxt('bpmpositionsCCL.txt',dtype = 'str',delimiter = ',')
keys = [i[-6:] for i in bpmpositions[:,0]]
values1 = np.array([float(i) for i in bpmpositions[:,1]])
values2 = np.array([float(i) for i in bpmpositions[:,2]])
values = values1+values2

bpmdict = dict(zip(keys, values))
```

```python
fig, ax = plt.subplots(figsize = (12,12))

surf = ax.imshow(np.abs(vmat),cmap=cm.cividis)

ax.tick_params(axis='both',labelsize=18)

ax.xaxis.set_ticks(np.arange(0,len(DCV_VA)-1,1))
ax.set_xticklabels(DCV_VA[1:],rotation =90)

ax.yaxis.set_ticks(np.arange(0,10,1))
ax.set_yticklabels(bpmdict.keys())

ax.set_title('Vertical Response Matrix', fontsize= 24)

cb = fig.colorbar(surf, ax=ax,shrink=0.45)

cb.ax.tick_params(labelsize = 20)
cb.ax.set_ylabel('Response Coefficient [mm/T]',fontsize = 20)

plt.show()
```

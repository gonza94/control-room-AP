```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'

from matplotlib import cbook, cm
from matplotlib.colors import LightSource
```

```python
DCH_VA=['DTL_Mag:PS_DCH618', 
'CCL_Mag:PS_DCH104',  
'CCL_Mag:PS_DCH106', 
'CCL_Mag:PS_DCH110', 
'CCL_Mag:PS_DCH112', 
'CCL_Mag:PS_DCH204',  
'CCL_Mag:PS_DCH206', 
'CCL_Mag:PS_DCH210', 
'CCL_Mag:PS_DCH212', 
'CCL_Mag:PS_DCH304', 
'CCL_Mag:PS_DCH306', 
'CCL_Mag:PS_DCH310', 
'CCL_Mag:PS_DCH312', 
'CCL_Mag:PS_DCH402', 
'CCL_Mag:PS_DCH404', 
'CCL_Mag:PS_DCH406', 
'CCL_Mag:PS_DCH408', 
'CCL_Mag:PS_DCH410']

DCH_VA = [s[-6:] for s in DCH_VA]
```

```python
hmat = np.loadtxt('orm_x.dat')

x = np.arange(0, hmat.shape[1])+1
y = np.arange(0, hmat.shape[0])+1
x, y = np.meshgrid(x, y)

bpmpositions = np.loadtxt('bpmpositionsCCL.txt',dtype = 'str',delimiter = ',')
keys = [i[-6:] for i in bpmpositions[:,0]]
values1 = np.array([float(i) for i in bpmpositions[:,1]])
values2 = np.array([float(i) for i in bpmpositions[:,2]])
values = values1+values2

bpmdict = dict(zip(keys, values))
```

```python
fig, ax = plt.subplots(figsize = (10,10),subplot_kw=dict(projection='3d'))

surf = ax.plot_surface(x, y, np.abs(hmat),cmap=cm.cividis)
 
plt.show()
```

```python
bottom = np.zeros_like(hmat.ravel())
width = depth = 1

fig, ax = plt.subplots(figsize = (10,10),subplot_kw=dict(projection='3d'))

cmap = cm.get_cmap('cividis')
rgba = [cmap((k)/np.max(hmat)) for k in np.abs(hmat.ravel())] 

ax.bar3d(x.ravel(), y.ravel(), bottom, width, depth, np.abs(hmat.ravel()), shade=True, color = rgba)
ax.set_title('Shaded')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,12))

surf = ax.imshow(np.abs(hmat),cmap=cm.cividis)

ax.tick_params(axis='both',labelsize=18)

ax.xaxis.set_ticks(np.arange(0,len(DCH_VA)-1,1))
ax.set_xticklabels(DCH_VA[1:],rotation =90)

ax.yaxis.set_ticks(np.arange(0,10,1))
ax.set_yticklabels(bpmdict.keys())

ax.set_title('Horizontal Response Matrix', fontsize= 24)

cb = fig.colorbar(surf, ax=ax,shrink=0.45)

cb.ax.tick_params(labelsize = 20)
cb.ax.set_ylabel('Response Coefficient [mm/T]',fontsize = 20)

plt.show()
```

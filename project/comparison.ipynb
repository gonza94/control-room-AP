```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'
```

```python
bare = np.loadtxt('before.txt')
lsqdata = np.loadtxt('lsqorbit.txt')
corrdata = np.loadtxt('after.txt')

x0 = np.loadtxt('xerror.txt')
y0 = np.loadtxt('yerror.txt')
xlsq = np.loadtxt('lsq.txt')
xlm = np.loadtxt('optimizerLM.txt')




bpm_pos = np.loadtxt('bpm_positions.txt',dtype = 'str',delimiter = ' ')
```

```python
bpm_pos[:,1].astype(float)
```

```python
fig,ax = plt.subplots(1,1,figsize = (10,6))

ax.plot(bare[:,0],1000*bare[:,1],'--',label = 'Bare Machine')
ax.plot(lsqdata[:,0],1000*lsqdata[:,1],label = 'LSQ')

ax.scatter(bpm_pos[:,1].astype(float),x0[1:], label = 'BPM Data (BM)')
ax.scatter(bpm_pos[:,1].astype(float),xlsq[1:,0], label = 'BPM Data (LSQ)')
#ax.plot(bpmdicts.keys(),datacorr[:,1]*1000,label = 'Corrected')

ax.set_xlabel('Longitudinal Position [m]',fontsize = 30)
ax.set_ylabel('Horizontal Orbit [mm]',fontsize = 30)

#ax.set_title('PyORBIT Tracking with Transfer Matrix Approach',fontsize = 20)

ax.tick_params(axis='both',labelsize=24)

#ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)

ax.legend(fontsize = 20)

plt.show()
plt.close()
```

```python
fig,ax = plt.subplots(1,1,figsize = (12,8))

ax.plot(bare[:,0],1000*bare[:,1],'--',label = 'Bare Machine')
ax.plot(lsqdata[:,0],1000*lsqdata[:,1],label = 'LSQ')
ax.plot(corrdata[:,0],1000*corrdata[:,1], label = 'LM')

ax.scatter(bpm_pos[:,1].astype(float),x0[1:], label = 'BPM Data (BM)')
ax.scatter(bpm_pos[:,1].astype(float),xlsq[1:,0], label = 'BPM Data (LSQ)')
ax.scatter(bpm_pos[:,1].astype(float),xlm[1:,0], label = 'BPM Data (LM)')
#ax.plot(bpmdicts.keys(),datacorr[:,1]*1000,label = 'Corrected')

ax.set_xlabel('Longitudinal Position [m]',fontsize = 30)
ax.set_ylabel('Horizontal Orbit [mm]',fontsize = 30)

#ax.set_title('PyORBIT Tracking with Transfer Matrix Approach',fontsize = 20)

ax.tick_params(axis='both',labelsize=24)

#ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)

ax.legend(loc = 3,fontsize = 20)

plt.show()
plt.close()
```

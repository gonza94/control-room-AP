```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'
```

```python
x0 = np.loadtxt('xnoerror.txt')
y0 = np.loadtxt('ynoerror.txt')

xnew = np.loadtxt('xerror.txt')
ynew = np.loadtxt('yerror.txt')
```

```python
bpmpositions = np.loadtxt('bpmpositionsCCL.txt',dtype = 'str',delimiter = ',')
keys = [i[-6:] for i in bpmpositions[:,0]]
values1 = np.array([float(i) for i in bpmpositions[:,1]])
values2 = np.array([float(i) for i in bpmpositions[:,2]])
values = values1+values2

bpmdict = dict(zip(keys, values))
```

```python
fig,ax = plt.subplots(1,1,figsize = (8,6))

ax.plot(bpmdict.keys(),x0,'--',label = 'Bare Machine')
ax.plot(bpmdict.keys(),xnew,label = 'With DTL Error')
#ax.plot(bpmdicts.keys(),datacorr[:,1]*1000,label = 'Corrected')

ax.set_xlabel('CCL BPMs',fontsize = 30)
ax.set_ylabel('Horizontal Orbit [mm]',fontsize = 30)

#ax.set_title('PyORBIT Tracking with Transfer Matrix Approach',fontsize = 20)

ax.tick_params(axis='both',labelsize=24)

ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)

ax.legend(fontsize = 20)

plt.show()
plt.close()
```

```python
fig,ax = plt.subplots(1,1,figsize = (8,6))

ax.plot(bpmdict.keys(),y0,'--',label = 'Bare Machine')
ax.plot(bpmdict.keys(),ynew,label = 'With DTL Error')
#ax.plot(bpmdicts.keys(),datacorr[:,1]*1000,label = 'Corrected')

ax.set_xlabel('CCL BPMs',fontsize = 30)
ax.set_ylabel('Vertical Orbit [mm]',fontsize = 30)

#ax.set_title('PyORBIT Tracking with Transfer Matrix Approach',fontsize = 20)

ax.tick_params(axis='both',labelsize=24)

ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)

ax.legend(fontsize = 20)

plt.show()
plt.close()
```

```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'
```

```python
# Solution to Problem 3 in Monday's homework
data = [[0.000,-85.9],
        [0.005,-53.6],
        [0.010,-21.0],
        [0.015,12.6],
        [0.020,46.2],
        [0.025,78.0],
        [0.030,111.6],
        [0.035,145.0],
        [0.040,178.4],
        [0.045,-147.9+360],
        [0.050,-114.9+360]]

data = np.array(data)
```

```python
fig,ax = plt.subplots(1,1, figsize = (8,6))

ax.scatter(data[:,0],data[:,1])

ax.set_xlabel('Actuator Position [m]',fontsize = 30)
ax.set_ylabel('BPM Phase [deg]',fontsize = 30)

ax.tick_params(axis='both',labelsize=24)

plt.show()
plt.close()
```

```python
fig,ax = plt.subplots(1,1, figsize = (8,6))

ax.scatter(data[:,0],data[:,1]-data[0,1])

ax.set_xlabel('Actuator Position [m]',fontsize = 30)
ax.set_ylabel('Phase Difference [deg]',fontsize = 30)

ax.tick_params(axis='both',labelsize=24)

plt.show()
plt.close()
```

```python
fig,ax = plt.subplots(1,1, figsize = (8,6))

ax.scatter(data[:,0],np.sin(data[:,1]*np.pi/180))

ax.set_xlabel('Actuator Position [m]',fontsize = 30)
ax.set_ylabel(r'$\sin \phi$',fontsize = 30)

ax.tick_params(axis='both',labelsize=24)

plt.show()
plt.close()
```

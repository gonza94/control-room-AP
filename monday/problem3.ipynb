```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'
```

```python
# Solution to Problem 3 in Monday's homework
data = [[0.000,-85.9],
        [0.005,-53.6],
        [0.010,-21.0],
        [0.015,12.6],
        [0.020,46.2],
        [0.025,78.0],
        [0.030,111.6],
        [0.035,145.0],
        [0.040,178.4],
        [0.045,-147.9+360],
        [0.050,-114.9+360]]

data = np.array(data)

z = np.polyfit(data[:,0], data[:,1], 1)
p = np.poly1d(z)
```

```python
z
```

```python
fig,ax = plt.subplots(1,1, figsize = (8,6))

ax.scatter(data[:,0],data[:,1],label = 'Data')

ax.plot(data[:,0],p(data[:,0]), 'r--',label = "$y=%.4f x- %.4f$"%(z[0],-z[1]))

ax.set_xlabel('Actuator Position [m]',fontsize = 30)
ax.set_ylabel('BPM Phase [deg]',fontsize = 30)

ax.tick_params(axis='both',labelsize=24)

ax.legend(fontsize = 20)

plt.show()
plt.close()
```

```python
c = 299792458 #[m/s]
f0 = 402.5e6 #Hz
mp = 938.27208816 #MeV/c^2
dzdphi = (1/z[0])*(180/np.pi)
beta = (2*np.pi*f0/c)*dzdphi
gamma = 1/np.sqrt(1-beta**2)
ek = (gamma-1)*mp #MeV
etotal = gamma*mp #MeV
```

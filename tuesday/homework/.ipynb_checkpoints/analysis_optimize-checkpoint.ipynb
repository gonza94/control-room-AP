```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'
```

```python
data = np.loadtxt('2024-01-30-20-00.csv',delimiter=',')
```

```python
pd.DataFrame(data, columns = ('DCV01','DCV04','DCV05','rms'))
```

```python
fig,ax = plt.subplots(1,1,figsize = (8,6))

ax.plot(data[:,0],data[:,1],label = 'DCV04 (Optimized)')
ax.plot(data[:,0],data[:,2],label = 'DCV05 (Optimized)')

ax.set_xlabel('DCV01 Strength [T/m]',fontsize = 30)
ax.set_ylabel('Corrector Strength [T/m]',fontsize = 30)

ax.tick_params(axis='both',labelsize=24)

ax.legend(fontsize = 20)

plt.show()
plt.close()
```

```python
databpm = np.loadtxt('bpmreadings.csv',delimiter=',')
databpmuncorr = np.array([-0.0021962,-0.56601,0.1905,0.42563,0.15599,-0.15075])
```

```python
pd.DataFrame(np.hstack((data,databpm)), columns = ('DCV01','DCV04','DCV05','rms','BPM01','BPM04','BPM05','BPM10','BPM11','BPM14'))
```

```python
fig,ax = plt.subplots(1,1,figsize = (8,6))

nbpms = ['BPM01','BPM04','BPM05','BPM10','BPM11','BPM14']

for i in np.arange(len(databpm)):
    ax.plot(nbpms,databpm[i], label = 'Corrected DCV01=%.3f T/m'%data[i,0])
ax.plot(nbpms,databpmuncorr,label = 'Uncorrected')

#ax.set_xlabel('BPM Number',fontsize = 30)
ax.set_ylabel('Vertical Orbit [mm]',fontsize = 30)

ax.tick_params(axis='both',labelsize=24)

ax.legend(fontsize = 14)

plt.show()
plt.close()
```

```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'
```

```python
data0 = np.loadtxt('uncorrected.csv',delimiter=',')
datacorr = np.loadtxt('corrected.csv',delimiter=',')

b1 = 0.01
b4 = 0.0009001391428
b5 = 0.0026138179
```

```python
fig,ax = plt.subplots(1,1,figsize = (8,6))

ax.plot(data0[:,0],data0[:,1]*1000,label = 'Uncorrected')
ax.plot(datacorr[:,0],datacorr[:,1]*1000,label = 'Corrected')

ax.set_xlabel('Longitudinal Position [m]',fontsize = 30)
ax.set_ylabel('Vertical Orbit [mm]',fontsize = 30)

ax.set_title('PyORBIT Tracking with Transfer Matrix Approach',fontsize = 20)

ax.tick_params(axis='both',labelsize=24)

ax.legend(fontsize = 20)

plt.show()
plt.close()
```
